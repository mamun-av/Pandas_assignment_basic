# load a new data file u.user and read it


```python
user = pd.read_csv("u.user", delimiter = "|")
user.head()
```

## Discover what is the mean age per occupation

```python
user.groupby(by = "occupation").agg({"age": "mean"})
```

## Discover the Male ratio per occupation and sort it from the most to the least


```python
male_count = user[user['gender']=='M'].groupby(by='occupation')['gender'].count()
total_count = user.groupby(by='occupation')['gender'].count()
ratio = (male_count/total_count).sort_values(ascending = False)
ratio
```

```python
new_table = pd.pivot_table(user,index='occupation',columns='gender',aggfunc='size')
new_table['total'] = new_table['M'] + new_table['F'] 
new_table['male_ratio'] = (new_table['M']/new_table['total'])*100
new_table.dropna()
new_table.sort_values(by=['male_ratio'],ascending=False)
```

## For each occupation, calculate the minimum and maximum ages

```python
user.groupby(by= "occupation").agg({"age": ["min", "max"]})
```

## For each combination of occupation and gender, calculate the mean age without heirarchical indexing

```python
user.groupby(["occupation", "gender"])["age"].mean()
```

## For each occupation present the percentage of women and men

```python
male_count = user[user['gender']=='M'].groupby(by='occupation')['gender'].count()
female_count = user[user['gender']=='F'].groupby(by='occupation')['gender'].count()

total_count = user.groupby(by='occupation')['gender'].count()  
```

```python
male_percentage = pd.DataFrame((male_count/total_count)*100)
female_percentage = pd.DataFrame((female_count/total_count)*100)

column_name = ['male','female']
final = pd.concat([male_percentage,female_percentage],axis=1)
final.columns = column_name
final
```

# Fictitious Names

```python
raw_data_1 = {
        'subject_id': ['1', '2', '3', '4', '5'],
        'first_name': ['Alex', 'Amy', 'Allen', 'Alice', 'Ayoung'], 
        'last_name': ['Anderson', 'Ackerman', 'Ali', 'Aoni', 'Atiches']}

raw_data_2 = {
        'subject_id': ['4', '5', '6', '7', '8'],
        'first_name': ['Billy', 'Brian', 'Bran', 'Bryce', 'Betty'], 
        'last_name': ['Bonder', 'Black', 'Balwner', 'Brice', 'Btisan']}

raw_data_3 = {
        'subject_id': ['1', '2', '3', '4', '5', '7', '8', '9', '10', '11'],
        'test_id': [51, 15, 15, 61, 16, 14, 15, 1, 61, 16]}
```

```python
data1 = pd.DataFrame(raw_data_1, columns = ['subject_id', 'first_name', 'last_name'])
data2 = pd.DataFrame(raw_data_2, columns = ['subject_id', 'first_name', 'last_name'])
data3 = pd.DataFrame(raw_data_3, columns = ['subject_id','test_id'])

data3
```

##  Join the two dataframes along rows and assign all_data

```python
all_data = pd.concat([data1, data2], axis=0, ignore_index = True)
all_data.head()
```

## Join the two dataframes along columns and assing to all_data_col

```python
all_data_col = pd.concat([data1, data2], axis = 1)
all_data_col
```

## Merge all_data and data3 along the subject_id value

```python
pd.merge(all_data, data3, on = "subject_id")
```

## Merge only the data that has the same 'subject_id' on both data1 and data2

```python
pd.merge(data1, data2, on = "subject_id")
```

## Merge all values in data1 and data2, with matching records from both sides where available.

```python
pd.merge(data1, data2, on = "subject_id", how = "inner")
```

```python
pd.merge(data1, data2, on = "subject_id", how = "outer")
```

# Wind data

## Assign it to a variable called data and replace the first 3 columns by a proper datetime index. 
 - dataset :wind.data

```python
import numpy as np
import pandas as pd
import datetime
```

```python
data = pd.read_fwf('wind.data', parse_dates = {"Date":["Yr","Mo","Dy"]}, keep_date_col=False, 
                   date_parser = lambda x,y,z: (datetime.datetime.strptime("19"+x+"-"+y+"-"+z, "%Y-%m-%d")))
```

```python
data.head()
```

##   Create a function to fix the year from 2061 to 1961  and apply it.

##  Set the right dates as the index. Pay attention at the data type, it should be datetime64[ns].

```python
data.set_index('Date', inplace = True)
```

```python
data.index
```

## Compute how many values are missing for each location over the entire record.¶

```python
data.head()
```

```python
data.isnull().sum()
```

## Compute how many non-missing values there are in total.

```python
sum(data.notnull().sum())
```

##  Calculate the mean windspeeds of the windspeeds over all the locations and all the times. ie a single number for enire dataset

```python
data.mean()
```

## Create a DataFrame called day_stats and calculate the min, max and mean windspeed and standard deviations of the windspeeds across all the locations at each day.


```python
day_stats = data.groupby("Date").agg(["min", "max","mean","std"])
day_stats
```

## Find the average windspeed in January for each location.



```python
data[data.index.month == 1].groupby(by='Date').mean()
```

## Downsample the record to a yearly frequency for each location.

```python
data
```

```python
yearly_resampled_data = data.resample('Y').mean()
yearly_resampled_data
```

## Downsample the record to a monthly frequency for each location.


```python
monthly_resampled_data = data.resample('M').mean()
monthly_resampled_data
```

## Downsample the record to a weekly frequency for each location.

```python
weekly_resampled_data = data.resample('W').count()
weekly_resampled_data
```

## Calculate the min, max and mean windspeeds and standard deviations of the windspeeds across all locations for each week (assume that the first week starts on January 2 1961) for the first 52 weeks.

```python
data.iloc[1:].resample('W').agg(['min','max','mean','std'])
```

## What is the year with most occurrences?

```python
data.resample('Y').sum().sum(axis=1).idxmax().year
```
